```python
import pandas as pd
import oracledb
from typing import List, Optional, Tuple
import traceback


def load_data_with_xml_and_error_tracking(total_records: int, batch_size: int = 1000, 
                                        user: str = None, password: str = None, dsn: str = None) -> Tuple[pd.DataFrame, List[str]]:
    """
    –ó–∞–≥—Ä—É–∂–∞–µ—Ç –¥–∞–Ω–Ω—ã–µ —Å XML –ø–∞—Ä—Å–∏–Ω–≥–æ–º –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –ø—Ä–æ–±–ª–µ–º–Ω—ã–µ itm_primary_key
    """
    
    all_data = []
    problematic_ids = []
    processed_records_count = 0
    
    try:
        conn = oracledb.connect(user=user, password=password, dsn=dsn)
        
        num_batches = (total_records + batch_size - 1) // batch_size
        
        print(f"–ó–∞–≥—Ä—É–∑–∫–∞ {total_records} –∑–∞–ø–∏—Å–µ–π –≤ {num_batches} –ø–∞–∫–µ—Ç–æ–≤...")
        
        for batch_num in range(num_batches):
            start_row = batch_num * batch_size + 1
            end_row = min((batch_num + 1) * batch_size, total_records)
            
            print(f"\n–ó–∞–≥—Ä—É–∑–∫–∞ –ø–∞–∫–µ—Ç–∞ {batch_num + 1}/{num_batches} (—Å—Ç—Ä–æ–∫–∏ {start_row}-{end_row})")
            
            try:
                # –ü—ã—Ç–∞–µ–º—Å—è –∑–∞–≥—Ä—É–∑–∏—Ç—å –≤–µ—Å—å –ø–∞–∫–µ—Ç —Å XML
                df_batch = load_batch_with_xml(start_row, end_row, conn)
                
                if df_batch is not None and len(df_batch) > 0:
                    all_data.append(df_batch)
                    processed_records_count += len(df_batch)
                    print(f"  ‚úÖ –£—Å–ø–µ—à–Ω–æ –∑–∞–≥—Ä—É–∂–µ–Ω–æ {len(df_batch)} –∑–∞–ø–∏—Å–µ–π —Å XML")
                else:
                    print(f"  ‚ÑπÔ∏è  –í –ø–∞–∫–µ—Ç–µ –Ω–µ—Ç –¥–∞–Ω–Ω—ã—Ö")
                
                progress = processed_records_count / total_records * 100
                print(f"  üìä –û–±—â–∏–π –ø—Ä–æ–≥—Ä–µ—Å—Å: {progress:.1f}% ({processed_records_count}/{total_records})")
                
            except Exception as batch_error:
                error_msg = str(batch_error)
                print(f"  ‚ùå –û—à–∏–±–∫–∞ XML –≤ –ø–∞–∫–µ—Ç–µ: {error_msg}")
                
                # –ù–∞—Ö–æ–¥–∏–º –ø—Ä–æ–±–ª–µ–º–Ω—É—é –∑–∞–ø–∏—Å—å –∏ –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ–º –ø–∞–∫–µ—Ç –ø–æ –æ–¥–Ω–æ–π –∑–∞–ø–∏—Å–∏
                successful_in_batch, problematic_in_batch = process_batch_record_by_record(start_row, end_row, conn)
                
                if successful_in_batch is not None and len(successful_in_batch) > 0:
                    all_data.append(successful_in_batch)
                    processed_records_count += len(successful_in_batch)
                    print(f"  ‚úÖ –ò–∑ –ø—Ä–æ–±–ª–µ–º–Ω–æ–≥–æ –ø–∞–∫–µ—Ç–∞ –¥–æ–±–∞–≤–ª–µ–Ω–æ {len(successful_in_batch)} —É—Å–ø–µ—à–Ω—ã—Ö –∑–∞–ø–∏—Å–µ–π")
                
                if problematic_in_batch:
                    problematic_ids.extend(problematic_in_batch)
                    print(f"  üö´ –ü—Ä–æ–±–ª–µ–º–Ω—ã–µ –∑–∞–ø–∏—Å–∏ –≤ –ø–∞–∫–µ—Ç–µ: {len(problematic_in_batch)}")
                
                print("  üîÑ –ü—Ä–æ–¥–æ–ª–∂–∞–µ–º —Å–æ —Å–ª–µ–¥—É—é—â–∏–º –ø–∞–∫–µ—Ç–æ–º...")
        
        # –û–±—ä–µ–¥–∏–Ω—è–µ–º –≤—Å–µ —É—Å–ø–µ—à–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
        if all_data:
            final_df = pd.concat(all_data, ignore_index=True)
            print(f"\nüéâ –ó–ê–ì–†–£–ó–ö–ê –ó–ê–í–ï–†–®–ï–ù–ê")
            print(f"‚úÖ –£—Å–ø–µ—à–Ω–æ –∑–∞–≥—Ä—É–∂–µ–Ω–æ: {len(final_df)} –∑–∞–ø–∏—Å–µ–π")
            print(f"üìä –û–±—Ä–∞–±–æ—Ç–∞–Ω–æ –∑–∞–ø–∏—Å–µ–π: {processed_records_count} –∏–∑ {total_records}")
            
            if problematic_ids:
                print(f"üö´ –ü—Ä–æ–±–ª–µ–º–Ω—ã–µ –∑–∞–ø–∏—Å–∏: {len(problematic_ids)}")
                print(f"üîç –ü–µ—Ä–≤—ã–µ 10 –ø—Ä–æ–±–ª–µ–º–Ω—ã—Ö itm_primary_key: {problematic_ids[:10]}")
            else:
                print("‚úÖ –í—Å–µ –∑–∞–ø–∏—Å–∏ –æ–±—Ä–∞–±–æ—Ç–∞–Ω—ã —É—Å–ø–µ—à–Ω–æ!")
                
            return final_df, problematic_ids
        else:
            print("‚ùå –ù–µ —É–¥–∞–ª–æ—Å—å –∑–∞–≥—Ä—É–∑–∏—Ç—å –Ω–∏ –æ–¥–Ω–æ–≥–æ –ø–∞–∫–µ—Ç–∞ –¥–∞–Ω–Ω—ã—Ö.")
            return pd.DataFrame(), problematic_ids
        
    except Exception as e:
        print(f"üí• –ö—Ä–∏—Ç–∏—á–µ—Å–∫–∞—è –æ—à–∏–±–∫–∞ –ø—Ä–∏ –∑–∞–≥—Ä—É–∑–∫–µ –¥–∞–Ω–Ω—ã—Ö: {e}")
        traceback.print_exc()
        return pd.DataFrame(), problematic_ids
    
    finally:
        if 'conn' in locals():
            conn.close()

def load_batch_with_xml(start_row: int, end_row: int, conn) -> pd.DataFrame:
    """
    –ó–∞–≥—Ä—É–∂–∞–µ—Ç –ø–∞–∫–µ—Ç –∑–∞–ø–∏—Å–µ–π —Å XML –ø–∞—Ä—Å–∏–Ω–≥–æ–º
    """
    try:
        cursor = conn.cursor()
        
        cursor.execute(base_sql, start_row=start_row, end_row=end_row)
        columns = [col[0] for col in cursor.description]
        data = cursor.fetchall()
        cursor.close()
        
        if data:
            return pd.DataFrame(data, columns=columns)
        else:
            return pd.DataFrame()
            
    except Exception as e:
        cursor.close()
        raise e

def process_batch_record_by_record(start_row: int, end_row: int, conn) -> Tuple[pd.DataFrame, List[str]]:
    """
    –û–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ—Ç –ø–∞–∫–µ—Ç –ø–æ –æ–¥–Ω–æ–π –∑–∞–ø–∏—Å–∏, –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —É—Å–ø–µ—à–Ω—ã–µ –∑–∞–ø–∏—Å–∏ –∏ –ø—Ä–æ–±–ª–µ–º–Ω—ã–µ ID
    """
    successful_records = []
    problematic_ids = []
    
    # –ü–æ–ª—É—á–∞–µ–º –≤—Å–µ ID –∑–∞–ø–∏—Å–µ–π –≤ –ø–∞–∫–µ—Ç–µ
    batch_ids = get_batch_record_ids(start_row, end_row, conn)
    
    if not batch_ids:
        return pd.DataFrame(), []
    
    print(f"  üîç –û–±—Ä–∞–±–æ—Ç–∫–∞ –ø–∞–∫–µ—Ç–∞ –ø–æ –æ–¥–Ω–æ–π –∑–∞–ø–∏—Å–∏ ({len(batch_ids)} –∑–∞–ø–∏—Å–µ–π)...")
    
    for i, record_id in enumerate(batch_ids):
        if i % 50 == 0:  # –ü—Ä–æ–≥—Ä–µ—Å—Å –∫–∞–∂–¥—ã–µ 50 –∑–∞–ø–∏—Å–µ–π
            print(f"    üìã –û–±—Ä–∞–±–æ—Ç–∞–Ω–æ {i}/{len(batch_ids)} –∑–∞–ø–∏—Å–µ–π...")
        
        try:
            # –ü—ã—Ç–∞–µ–º—Å—è –∑–∞–≥—Ä—É–∑–∏—Ç—å –∑–∞–ø–∏—Å—å —Å XML
            df_single = load_single_record_with_xml(record_id, conn)
            if df_single is not None and len(df_single) > 0:
                successful_records.append(df_single)
        except Exception as e:
            # –ï—Å–ª–∏ –æ—à–∏–±–∫–∞ - –¥–æ–±–∞–≤–ª—è–µ–º –≤ –ø—Ä–æ–±–ª–µ–º–Ω—ã–µ
            problematic_ids.append(record_id)
            print(f"    üö´ –û—à–∏–±–∫–∞ –≤ –∑–∞–ø–∏—Å–∏ {record_id}: {str(e)}")
    
    # –û–±—ä–µ–¥–∏–Ω—è–µ–º –≤—Å–µ —É—Å–ø–µ—à–Ω—ã–µ –∑–∞–ø–∏—Å–∏
    if successful_records:
        successful_df = pd.concat(successful_records, ignore_index=True)
        return successful_df, problematic_ids
    else:
        return pd.DataFrame(), problematic_ids

def get_batch_record_ids(start_row: int, end_row: int, conn) -> List[str]:
    """
    –ü–æ–ª—É—á–∞–µ—Ç —Å–ø–∏—Å–æ–∫ itm_primary_key –¥–ª—è –ø–∞–∫–µ—Ç–∞
    """
    try:
        cursor = conn.cursor()
        
        id_query = """
        WITH suppliers AS (
            SELECT it.itm_primary_key,
                   ROW_NUMBER() OVER (ORDER BY it.ITM_LAST_MODIFIED) as rn
            FROM tctg_itm_item it
            LEFT JOIN tctg_itx_item_content ct ON ct.itx_entry_id = it.ITM_ID
            WHERE (it.itm_container_id = '1840' OR it.itm_container_id = '1839')
        )
        SELECT itm_primary_key
        FROM suppliers
        WHERE rn BETWEEN :start_row AND :end_row
        ORDER BY rn
        """
        
        cursor.execute(id_query, start_row=start_row, end_row=end_row)
        batch_ids = [row[0] for row in cursor.fetchall()]
        cursor.close()
        
        return batch_ids
    except Exception as e:
        print(f"    ‚ùå –û—à–∏–±–∫–∞ –ø—Ä–∏ –ø–æ–ª—É—á–µ–Ω–∏–∏ ID –ø–∞–∫–µ—Ç–∞: {e}")
        return []

def load_single_record_with_xml(record_id: str, conn) -> pd.DataFrame:
    """
    –ó–∞–≥—Ä—É–∂–∞–µ—Ç –æ–¥–Ω—É –∑–∞–ø–∏—Å—å —Å –ø–æ–ª–Ω—ã–º XML –ø–∞—Ä—Å–∏–Ω–≥–æ–º
    """
    try:
        cursor = conn.cursor()
        
        # –ò–°–ü–†–ê–í–õ–ï–ù–ù–´–ô –ó–ê–ü–†–û–° - —É–±—Ä–∞–ª –ø—Å–µ–≤–¥–æ–Ω–∏–º "it."

        
        cursor.execute(single_sql, record_id=record_id)
        columns = [col[0] for col in cursor.description]
        data = cursor.fetchall()
        cursor.close()
        
        if data:
            return pd.DataFrame(data, columns=columns)
        else:
            return pd.DataFrame()
            
    except Exception as e:
        if 'cursor' in locals():
            cursor.close()
        raise e

# –£–ø—Ä–æ—â–µ–Ω–Ω–∞—è –≤–µ—Ä—Å–∏—è –¥–ª—è —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è
def load_data_simple_fallback(total_records: int, batch_size: int = 1000, 
                            user: str = None, password: str = None, dsn: str = None) -> Tuple[pd.DataFrame, List[str]]:
    """
    –£–ø—Ä–æ—â–µ–Ω–Ω–∞—è –∑–∞–≥—Ä—É–∑–∫–∞ —Ç–æ–ª—å–∫–æ –±–∞–∑–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö –±–µ–∑ XML
    """
    print("üîÑ –ò—Å–ø–æ–ª—å–∑—É–µ–º —É–ø—Ä–æ—â–µ–Ω–Ω—É—é –∑–∞–≥—Ä—É–∑–∫—É (–±–µ–∑ XML)...")
    
    simple_sql = """
    WITH suppliers AS (
        SELECT it.itm_primary_key,
               it.ITM_LAST_MODIFIED,
               it.ITM_NEXT_VERSION_ID,
               ct.ITX_CONTAINER_ID,
               ROW_NUMBER() OVER (ORDER BY it.ITM_LAST_MODIFIED) as rn
        FROM tctg_itm_item it
        LEFT JOIN tctg_itx_item_content ct ON ct.itx_entry_id = it.ITM_ID
        WHERE (it.itm_container_id = '1840' OR it.itm_container_id = '1839')
    )
    SELECT itm_primary_key,
           ITM_LAST_MODIFIED,
           ITM_NEXT_VERSION_ID,
           ITX_CONTAINER_ID
    FROM suppliers
    WHERE rn BETWEEN :start_row AND :end_row
    ORDER BY rn
    """
    
    all_data = []
    
    try:
        conn = oracledb.connect(user=user, password=password, dsn=dsn)
        cursor = conn.cursor()
        
        num_batches = (total_records + batch_size - 1) // batch_size
        
        for batch_num in range(num_batches):
            start_row = batch_num * batch_size + 1
            end_row = min((batch_num + 1) * batch_size, total_records)
            
            print(f"–ü–∞–∫–µ—Ç {batch_num + 1}/{num_batches}...")
            
            cursor.execute(simple_sql, start_row=start_row, end_row=end_row)
            columns = [col[0] for col in cursor.description]
            data = cursor.fetchall()
            
            if data:
                df_batch = pd.DataFrame(data, columns=columns)
                all_data.append(df_batch)
                print(f"  –ó–∞–≥—Ä—É–∂–µ–Ω–æ {len(df_batch)} –∑–∞–ø–∏—Å–µ–π")
        
        cursor.close()
        
        if all_data:
            final_df = pd.concat(all_data, ignore_index=True)
            print(f"‚úÖ –£—Å–ø–µ—à–Ω–æ –∑–∞–≥—Ä—É–∂–µ–Ω–æ: {len(final_df)} –∑–∞–ø–∏—Å–µ–π (–±–µ–∑ XML)")
            return final_df, []
        else:
            return pd.DataFrame(), []
            
    except Exception as e:
        print(f"‚ùå –û—à–∏–±–∫–∞: {e}")
        return pd.DataFrame(), []
    finally:
        if 'conn' in locals():
            conn.close()

```

```python
def replace_headers_with_nod_names(df_raw, attr):
    """
    –ó–∞–º–µ–Ω—è–µ—Ç –≤ –∑–∞–≥–æ–ª–æ–≤–∫–∞—Ö df_raw –≤—Å–µ –ø–æ–¥—Å—Ç—Ä–æ–∫–∏, —Å–æ–≤–ø–∞–¥–∞—é—â–∏–µ —Å NOD_ID, –Ω–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–µ NOD_NAME –∏–∑ attr
    """
    # –°–æ–∑–¥–∞–µ–º —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –∑–∞–º–µ–Ω—ã: NOD_ID -> NOD_NAME
    replacement_dict = {}
    
    # –ó–∞–ø–æ–ª–Ω—è–µ–º —Å–ª–æ–≤–∞—Ä—å –∏–∑ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ attr
    for index, row in attr.iterrows():
        nod_id = str(row['NOD_ID'])
        nod_name = str(row['NOD_NAME'])
        replacement_dict[nod_id] = nod_name
    
    print(f"üîß –°–ª–æ–≤–∞—Ä—å –∑–∞–º–µ–Ω—ã —Å–æ–¥–µ—Ä–∂–∏—Ç {len(replacement_dict)} —ç–ª–µ–º–µ–Ω—Ç–æ–≤")
    print("üìã –ü—Ä–∏–º–µ—Ä—ã –∑–∞–º–µ–Ω:")
    for i, (nod_id, nod_name) in enumerate(list(replacement_dict.items())[:5]):
        print(f"   {nod_id} -> {nod_name}")
    
    # –°–æ–∑–¥–∞–µ–º –∫–æ–ø–∏—é –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ —á—Ç–æ–±—ã –Ω–µ –∏–∑–º–µ–Ω—è—Ç—å –æ—Ä–∏–≥–∏–Ω–∞–ª
    df_modified = df_raw.copy()
    
    # –ù–æ–≤—ã–µ –Ω–∞–∑–≤–∞–Ω–∏—è –∫–æ–ª–æ–Ω–æ–∫
    new_columns = []
    
    # –û–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ–º –∫–∞–∂–¥—É—é –∫–æ–ª–æ–Ω–∫—É
    for column in df_raw.columns:
        new_column = str(column)
        
        # –ó–∞–º–µ–Ω—è–µ–º –≤—Å–µ –≤—Ö–æ–∂–¥–µ–Ω–∏—è NOD_ID –Ω–∞ NOD_NAME
        for nod_id, nod_name in replacement_dict.items():
            if nod_id in new_column:
                new_column = new_column.replace(nod_id, nod_name)
                print(f"üîÑ –í –∫–æ–ª–æ–Ω–∫–µ '{column}' –∑–∞–º–µ–Ω–µ–Ω–æ '{nod_id}' –Ω–∞ '{nod_name}'")
        
        new_columns.append(new_column)
    
    # –ü—Ä–∏–º–µ–Ω—è–µ–º –Ω–æ–≤—ã–µ –Ω–∞–∑–≤–∞–Ω–∏—è –∫–æ–ª–æ–Ω–æ–∫
    df_modified.columns = new_columns
    
    print(f"‚úÖ –ó–∞–º–µ–Ω–∞ –∑–∞–≤–µ—Ä—à–µ–Ω–∞. –û–±—Ä–∞–±–æ—Ç–∞–Ω–æ {len(df_raw.columns)} –∫–æ–ª–æ–Ω–æ–∫")
    
    # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º –∏–∑–º–µ–Ω–µ–Ω–∏—è
    print("\nüìä –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –∑–∞–≥–æ–ª–æ–≤–∫–æ–≤:")
    for old_col, new_col in zip(df_raw.columns, df_modified.columns):
        if old_col != new_col:
            print(f"   '{old_col}' -> '{new_col}'")
    
    return df_modified

# –ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≤–µ—Ä—Å–∏—è —Å —É—á–µ—Ç–æ–º —á–∞—Å—Ç–∏—á–Ω—ã—Ö —Å–æ–≤–ø–∞–¥–µ–Ω–∏–π
def replace_headers_smart(df_raw, attr):
    """
    –£–º–Ω–∞—è –∑–∞–º–µ–Ω–∞ —Å —É—á–µ—Ç–æ–º —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —Ñ–æ—Ä–º–∞—Ç–æ–≤ NOD_ID –≤ –∑–∞–≥–æ–ª–æ–≤–∫–∞—Ö
    """
    # –°–æ–∑–¥–∞–µ–º —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –∑–∞–º–µ–Ω—ã
    replacement_dict = {}
    
    for index, row in attr.iterrows():
        nod_id = str(row['NOD_ID']).strip()
        nod_name = str(row['NOD_NAME']).strip()
        if nod_id and nod_name:  # –ü—Ä–æ–≤–µ—Ä—è–µ–º —á—Ç–æ –∑–Ω–∞—á–µ–Ω–∏—è –Ω–µ –ø—É—Å—Ç—ã–µ
            replacement_dict[nod_id] = nod_name
    
    print(f"üîß –ù–∞–π–¥–µ–Ω–æ {len(replacement_dict)} –ø–∞—Ä –¥–ª—è –∑–∞–º–µ–Ω—ã")
    
    # –°–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ –¥–ª–∏–Ω–µ ID (–æ—Ç —Å–∞–º—ã—Ö –¥–ª–∏–Ω–Ω—ã—Ö –∫ —Å–∞–º—ã–º –∫–æ—Ä–æ—Ç–∫–∏–º)
    # —á—Ç–æ–±—ã —Å–Ω–∞—á–∞–ª–∞ –∑–∞–º–µ–Ω—è—Ç—å –¥–ª–∏–Ω–Ω—ã–µ —Å–æ–≤–ø–∞–¥–µ–Ω–∏—è
    sorted_replacements = sorted(replacement_dict.items(), key=lambda x: len(x[0]), reverse=True)
    
    df_modified = df_raw.copy()
    new_columns = []
    changes_made = 0
    
    for column in df_raw.columns:
        original_column = str(column)
        modified_column = original_column
        
        # –ü—Ä–æ–±—É–µ–º –≤—Å–µ –≤–æ–∑–º–æ–∂–Ω—ã–µ –∑–∞–º–µ–Ω—ã
        for nod_id, nod_name in sorted_replacements:
            if nod_id in modified_column:
                modified_column = modified_column.replace(nod_id, nod_name)
                changes_made += 1
                print(f"üîÑ –ó–∞–º–µ–Ω–∞ –≤ '{original_column}': '{nod_id}' -> '{nod_name}'")
        
        new_columns.append(modified_column)
    
    df_modified.columns = new_columns
    
    print(f"‚úÖ –í—ã–ø–æ–ª–Ω–µ–Ω–æ {changes_made} –∑–∞–º–µ–Ω –≤ {len(df_raw.columns)} –∫–æ–ª–æ–Ω–∫–∞—Ö")
    
    return df_modified

# –í–µ—Ä—Å–∏—è —Å —É—á–µ—Ç–æ–º —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —Ñ–æ—Ä–º–∞—Ç–æ–≤ (—Ü–∏—Ñ—Ä—ã, —Å—Ç—Ä–æ–∫–∏ –∏ —Ç.–¥.)
def replace_headers_comprehensive(df_raw, attr):
    """
    –ö–æ–º–ø–ª–µ–∫—Å–Ω–∞—è –∑–∞–º–µ–Ω–∞ —Å —É—á–µ—Ç–æ–º —Ä–∞–∑–Ω—ã—Ö —Ñ–æ—Ä–º–∞—Ç–æ–≤ –¥–∞–Ω–Ω—ã—Ö
    """
    # –°–æ–∑–¥–∞–µ–º —Ä–∞—Å—à–∏—Ä–µ–Ω–Ω—ã–π —Å–ª–æ–≤–∞—Ä—å –∑–∞–º–µ–Ω
    replacement_dict = {}
    
    for index, row in attr.iterrows():
        nod_id = str(row['NOD_ID']).strip()
        nod_name = str(row['NOD_NAME']).strip()
        
        if nod_id and nod_name:
            # –î–æ–±–∞–≤–ª—è–µ–º —Ä–∞–∑–ª–∏—á–Ω—ã–µ –≤–∞—Ä–∏–∞–Ω—Ç—ã ID
            replacement_dict[nod_id] = nod_name
            
            # –ï—Å–ª–∏ ID - —á–∏—Å–ª–æ, –¥–æ–±–∞–≤–ª—è–µ–º –µ–≥–æ —Å—Ç—Ä–æ–∫–æ–≤–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ
            try:
                if nod_id.isdigit():
                    # –î–æ–±–∞–≤–ª—è–µ–º –≤–∞—Ä–∏–∞–Ω—Ç—ã —Å —Ä–∞–∑–Ω—ã–º–∏ —Ñ–æ—Ä–º–∞—Ç–∞–º–∏
                    replacement_dict[f"_{nod_id}"] = f"_{nod_name}"
                    replacement_dict[f"{nod_id}_"] = f"{nod_name}_"
            except:
                pass
    
    print(f"üîß –†–∞—Å—à–∏—Ä–µ–Ω–Ω—ã–π —Å–ª–æ–≤–∞—Ä—å —Å–æ–¥–µ—Ä–∂–∏—Ç {len(replacement_dict)} —ç–ª–µ–º–µ–Ω—Ç–æ–≤")
    
    df_modified = df_raw.copy()
    new_columns = []
    
    for column in df_raw.columns:
        column_str = str(column)
        new_column = column_str
        
        # –°–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ –¥–ª–∏–Ω–µ (—Å–Ω–∞—á–∞–ª–∞ –¥–ª–∏–Ω–Ω—ã–µ —Å–æ–≤–ø–∞–¥–µ–Ω–∏—è)
        for nod_id, nod_name in sorted(replacement_dict.items(), key=lambda x: len(x[0]), reverse=True):
            if nod_id in new_column:
                new_column = new_column.replace(nod_id, nod_name)
        
        new_columns.append(new_column)
    
    df_modified.columns = new_columns
    
    # –°–æ–∑–¥–∞–µ–º –æ—Ç—á–µ—Ç –æ –∏–∑–º–µ–Ω–µ–Ω–∏—è—Ö
    changed_columns = [(old, new) for old, new in zip(df_raw.columns, df_modified.columns) if old != new]
    
    print(f"‚úÖ –ò–∑–º–µ–Ω–µ–Ω–æ {len(changed_columns)} –∏–∑ {len(df_raw.columns)} –∫–æ–ª–æ–Ω–æ–∫")
    if changed_columns:
        print("üìã –ò–∑–º–µ–Ω–µ–Ω–Ω—ã–µ –∫–æ–ª–æ–Ω–∫–∏:")
        for old, new in changed_columns:
            print(f"   '{old}' -> '{new}'")
    
    return df_modified

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
def check_replacement_results(original_df, modified_df, attr):
    """
    –ü—Ä–æ–≤–µ—Ä—è–µ—Ç —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –∑–∞–º–µ–Ω—ã –∏ –≤—ã–≤–æ–¥–∏—Ç —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É
    """
    print("\n" + "="*50)
    print("üìä –û–¢–ß–ï–¢ –û –ó–ê–ú–ï–ù–ï –ó–ê–ì–û–õ–û–í–ö–û–í")
    print("="*50)
    
    total_columns = len(original_df.columns)
    changed_columns = 0
    
    for i, (old_col, new_col) in enumerate(zip(original_df.columns, modified_df.columns)):
        if old_col != new_col:
            changed_columns += 1
            print(f"{i+1:2d}. ‚úÖ –ò–ó–ú–ï–ù–ï–ù–û: '{old_col}' -> '{new_col}'")
        else:
            print(f"{i+1:2d}. ‚è© –ë–ï–ó –ò–ó–ú–ï–ù–ï–ù–ò–ô: '{old_col}'")
    
    print(f"\nüìà –ò–¢–û–ì–û: {changed_columns} –∏–∑ {total_columns} –∫–æ–ª–æ–Ω–æ–∫ –∏–∑–º–µ–Ω–µ–Ω–æ")
    
    # –ü—Ä–æ–≤–µ—Ä—è–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ NOD_NAME –∏–∑ attr
    used_nod_names = set()
    for nod_name in attr['NOD_NAME']:
        nod_name_str = str(nod_name)
        for column in modified_df.columns:
            if nod_name_str in column:
                used_nod_names.add(nod_name_str)
    
    print(f"üî§ –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö NOD_NAME: {len(used_nod_names)} –∏–∑ {len(attr)}")

# –û—Å–Ω–æ–≤–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è
def process_dataframe_headers(df_raw, attr, method='smart'):
    """
    –û—Å–Ω–æ–≤–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏ –∑–∞–≥–æ–ª–æ–≤–∫–æ–≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞
    
    Parameters:
    df_raw: –∏—Å—Ö–æ–¥–Ω—ã–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º
    attr: –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å mapping (NOD_ID, NOD_NAME)
    method: –º–µ—Ç–æ–¥ –∑–∞–º–µ–Ω—ã ('simple', 'smart', 'comprehensive')
    """
    print("üéØ –ù–ê–ß–ê–õ–û –û–ë–†–ê–ë–û–¢–ö–ò –ó–ê–ì–û–õ–û–í–ö–û–í")
    print(f"üìÅ –ò—Å—Ö–æ–¥–Ω—ã–µ –∫–æ–ª–æ–Ω–∫–∏: {len(df_raw.columns)}")
    print(f"üìä –î–∞–Ω–Ω—ã–µ –¥–ª—è –∑–∞–º–µ–Ω—ã: {len(attr)} –∑–∞–ø–∏—Å–µ–π")
    
    if method == 'simple':
        result_df = replace_headers_with_nod_names(df_raw, attr)
    elif method == 'smart':
        result_df = replace_headers_smart(df_raw, attr)
    elif method == 'comprehensive':
        result_df = replace_headers_comprehensive(df_raw, attr)
    else:
        print("‚ùå –ù–µ–∏–∑–≤–µ—Å—Ç–Ω—ã–π –º–µ—Ç–æ–¥. –ò—Å–ø–æ–ª—å–∑—É—é 'smart'")
        result_df = replace_headers_smart(df_raw, attr)
    
    # –ü—Ä–æ–≤–µ—Ä—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
    check_replacement_results(df_raw, result_df, attr)
    
    return result_df
```

```python
import pandas as pd
import oracledb

def save_with_correct_encoding(df_raw, filename="data.xlsx"):
    """
    –°–æ—Ö—Ä–∞–Ω—è–µ—Ç —Å –ø—Ä–∞–≤–∏–ª—å–Ω–æ–π –∫–æ–¥–∏—Ä–æ–≤–∫–æ–π –¥–ª—è –∫–∏—Ä–∏–ª–ª–∏—Ü—ã –≤ Excel
    """
    print("üîÑ –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —Å –ø—Ä–∞–≤–∏–ª—å–Ω–æ–π –∫–æ–¥–∏—Ä–æ–≤–∫–æ–π...")
    
    try:
        # –ò–∑–≤–ª–µ–∫–∞–µ–º –¥–∞–Ω–Ω—ã–µ –∏–∑ CLOB –æ–±—ä–µ–∫—Ç–æ–≤
        extracted_data = []
        
        for index in range(len(df_raw)):
            if index % 1000 == 0:
                print(f"üìä –û–±—Ä–∞–±–æ—Ç–∞–Ω–æ {index} —Å—Ç—Ä–æ–∫...")
            
            row_data = {}
            for col in df_raw.columns:
                value = df_raw.iloc[index][col]
                
                # –û–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ–º CLOB –æ–±—ä–µ–∫—Ç—ã
                if hasattr(value, 'read'):
                    try:
                        clob_content = value.read()
                        if clob_content:
                            row_data[col] = str(clob_content)
                        else:
                            row_data[col] = ""
                    except:
                        row_data[col] = "CLOB_ERROR"
                else:
                    row_data[col] = value
            
            extracted_data.append(row_data)
        
        # –°–æ–∑–¥–∞–µ–º —á–∏—Å—Ç—ã–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º
        clean_df = pd.DataFrame(extracted_data)
        
        # –°–æ—Ö—Ä–∞–Ω—è–µ–º –≤ Excel (–∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã–π —Å–∏–Ω—Ç–∞–∫—Å–∏—Å)
        clean_df.to_excel(filename, index=False, engine='openpyxl')
        
        print(f"‚úÖ –§–∞–π–ª —Å–æ—Ö—Ä–∞–Ω–µ–Ω: {filename}")
        return True
        
    except Exception as e:
        print(f"‚ùå –û—à–∏–±–∫–∞: {e}")
        return False

def save_with_encoding_fix(df_raw, filename="data_fixed.xlsx"):
    """
    –°–æ—Ö—Ä–∞–Ω—è–µ—Ç —Å –ø—Ä–∏–Ω—É–¥–∏—Ç–µ–ª—å–Ω–æ–π –æ–±—Ä–∞–±–æ—Ç–∫–æ–π –∫–∏—Ä–∏–ª–ª–∏—Ü—ã
    """
    print("üîÑ –ü—Ä–∏–Ω—É–¥–∏—Ç–µ–ª—å–Ω–∞—è –æ–±—Ä–∞–±–æ—Ç–∫–∞ –∫–∏—Ä–∏–ª–ª–∏—Ü—ã...")
    
    try:
        # –û–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ–º –¥–∞–Ω–Ω—ã–µ —Å —É—á–µ—Ç–æ–º –∫–æ–¥–∏—Ä–æ–≤–∫–∏
        processed_data = []
        
        for index in range(len(df_raw)):
            row_data = {}
            for col in df_raw.columns:
                value = df_raw.iloc[index][col]
                
                if value is None:
                    row_data[col] = ""
                elif hasattr(value, 'read'):
                    try:
                        content = value.read()
                        if content:
                            if isinstance(content, bytes):
                                # –ü—Ä–æ–±—É–µ–º —Ä–∞–∑–Ω—ã–µ –∫–æ–¥–∏—Ä–æ–≤–∫–∏ –¥–ª—è –∫–∏—Ä–∏–ª–ª–∏—Ü—ã
                                try:
                                    row_data[col] = content.decode('utf-8')
                                except UnicodeDecodeError:
                                    try:
                                        row_data[col] = content.decode('cp1251')
                                    except UnicodeDecodeError:
                                        try:
                                            row_data[col] = content.decode('iso-8859-1')
                                        except UnicodeDecodeError:
                                            row_data[col] = content.decode('latin1')
                            else:
                                row_data[col] = str(content)
                        else:
                            row_data[col] = ""
                    except Exception as e:
                        row_data[col] = f"–û—à–∏–±–∫–∞ —á—Ç–µ–Ω–∏—è: {e}"
                else:
                    row_data[col] = str(value) if value is not None else ""
            
            processed_data.append(row_data)
        
        # –°–æ–∑–¥–∞–µ–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º
        df_fixed = pd.DataFrame(processed_data)
        
        # –°–æ—Ö—Ä–∞–Ω—è–µ–º
        df_fixed.to_excel(filename, index=False, engine='openpyxl')
        
        print(f"‚úÖ –§–∞–π–ª —Å –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–æ–π –∫–æ–¥–∏—Ä–æ–≤–∫–æ–π: {filename}")
        return True
        
    except Exception as e:
        print(f"‚ùå –û—à–∏–±–∫–∞: {e}")
        return False

def save_as_csv_first(df_raw, csv_filename="data_utf8.csv", excel_filename="data.xlsx"):
    """
    –°–æ—Ö—Ä–∞–Ω—è–µ—Ç —Å–Ω–∞—á–∞–ª–∞ –≤ CSV —Å UTF-8 BOM, –ø–æ—Ç–æ–º –∫–æ–Ω–≤–µ—Ä—Ç–∏—Ä—É–µ—Ç –≤ Excel
    """
    print("üîÑ –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —á–µ—Ä–µ–∑ CSV —Å UTF-8 BOM...")
    
    try:
        # –û–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ–º –¥–∞–Ω–Ω—ã–µ
        processed_data = []
        
        for index in range(len(df_raw)):
            row_data = {}
            for col in df_raw.columns:
                value = df_raw.iloc[index][col]
                
                if hasattr(value, 'read'):
                    try:
                        content = value.read()
                        row_data[col] = content if content else ""
                    except:
                        row_data[col] = ""
                else:
                    row_data[col] = value if value is not None else ""
            
            processed_data.append(row_data)
        
        df_clean = pd.DataFrame(processed_data)
        
        # –°–æ—Ö—Ä–∞–Ω—è–µ–º –≤ CSV —Å UTF-8 BOM (–¥–ª—è Excel)
        df_clean.to_csv(csv_filename, index=False, encoding='utf-8-sig', sep=',')
        print(f"‚úÖ CSV —Å BOM —Å–æ—Ö—Ä–∞–Ω–µ–Ω: {csv_filename}")
        
        # –ß–∏—Ç–∞–µ–º CSV –∏ —Å–æ—Ö—Ä–∞–Ω—è–µ–º –≤ Excel
        df_from_csv = pd.read_csv(csv_filename, encoding='utf-8-sig', sep=',')
        df_from_csv.to_excel(excel_filename, index=False, engine='openpyxl')
        print(f"‚úÖ Excel —Ñ–∞–π–ª —Å–æ–∑–¥–∞–Ω: {excel_filename}")
        
        return True
        
    except Exception as e:
        print(f"‚ùå –û—à–∏–±–∫–∞: {e}")
        return False

def ultimate_save_solution(df_raw, filename="ultimate_data.xlsx"):
    """
    –£–ª—å—Ç–∏–º–∞—Ç–∏–≤–Ω–æ–µ —Ä–µ—à–µ–Ω–∏–µ –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è —Å –∫–∏—Ä–∏–ª–ª–∏—Ü–µ–π
    """
    print("üöÄ –ó–ê–ü–£–°–ö –£–õ–¨–¢–ò–ú–ê–¢–ò–í–ù–û–ì–û –°–û–•–†–ê–ù–ï–ù–ò–Ø")
    
    try:
        # –ú–µ—Ç–æ–¥ 1: –ì–ª—É–±–æ–∫–æ–µ –∫–æ–ø–∏—Ä–æ–≤–∞–Ω–∏–µ —Å –æ–±—Ä–∞–±–æ—Ç–∫–æ–π –∫–æ–¥–∏—Ä–æ–≤–∫–∏
        print("1. –ì–ª—É–±–æ–∫–æ–µ –∫–æ–ø–∏—Ä–æ–≤–∞–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö...")
        
        all_data = []
        for i in range(len(df_raw)):
            if i % 1000 == 0:
                print(f"   –û–±—Ä–∞–±–æ—Ç–∞–Ω–æ {i} —Å—Ç—Ä–æ–∫...")
                
            row_dict = {}
            for col in df_raw.columns:
                original_value = df_raw.iloc[i][col]
                
                # –û–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ–º CLOB
                if hasattr(original_value, 'read'):
                    try:
                        clob_content = original_value.read()
                        if clob_content:
                            # –ü—Ä–∏–Ω—É–¥–∏—Ç–µ–ª—å–Ω–æ –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ–º –∫–∞–∫ —Å—Ç—Ä–æ–∫—É
                            if isinstance(clob_content, bytes):
                                # –ü—Ä–æ–±—É–µ–º —Ä–∞–∑–Ω—ã–µ –∫–æ–¥–∏—Ä–æ–≤–∫–∏ –¥–ª—è –∫–∏—Ä–∏–ª–ª–∏—Ü—ã
                                encodings = ['utf-8', 'cp1251', 'iso-8859-1', 'latin1']
                                for encoding in encodings:
                                    try:
                                        row_dict[col] = clob_content.decode(encoding)
                                        break
                                    except UnicodeDecodeError:
                                        continue
                                else:
                                    row_dict[col] = "DECODING_ERROR"
                            else:
                                row_dict[col] = str(clob_content)
                        else:
                            row_dict[col] = ""
                    except Exception as e:
                        row_dict[col] = f"–û—à–∏–±–∫–∞: {e}"
                else:
                    # –û–±—ã—á–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è
                    if original_value is None:
                        row_dict[col] = ""
                    else:
                        row_dict[col] = original_value
            
            all_data.append(row_dict)
        
        # –°–æ–∑–¥–∞–µ–º –Ω–æ–≤—ã–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º
        ultimate_df = pd.DataFrame(all_data)
        
        print("2. –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –≤ Excel...")
        
        # –ü—Ä–æ—Å—Ç–æ–µ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –±–µ–∑ –ª–∏—à–Ω–∏—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
        ultimate_df.to_excel(filename, index=False, engine='openpyxl')
        
        print(f"‚úÖ –£–õ–¨–¢–ò–ú–ê–¢–ò–í–ù–û–ï –°–û–•–†–ê–ù–ï–ù–ò–ï –ó–ê–í–ï–†–®–ï–ù–û: {filename}")
        print("üî§ –ö–ò–†–ò–õ–õ–ò–¶–ê –î–û–õ–ñ–ù–ê –û–¢–û–ë–†–ê–ñ–ê–¢–¨–°–Ø –ö–û–†–†–ï–ö–¢–ù–û!")
        
        return True
        
    except Exception as e:
        print(f"‚ùå –£–ª—å—Ç–∏–º–∞—Ç–∏–≤–Ω–æ–µ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –Ω–µ —É–¥–∞–ª–æ—Å—å: {e}")
        return False

def simple_russian_save(df_raw, filename="simple_russian.xlsx"):
    """
    –ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –ø—Ä–æ—Å—Ç–æ–µ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —Å –æ–±—Ä–∞–±–æ—Ç–∫–æ–π –∫–∏—Ä–∏–ª–ª–∏—Ü—ã
    """
    print("üîÑ –ü—Ä–æ—Å—Ç–æ–µ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –¥–ª—è —Ä—É—Å—Å–∫–æ–≥–æ —è–∑—ã–∫–∞...")
    
    try:
        # –ü—Ä–æ—Å—Ç–æ –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ–º –≤—Å–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ —Å—Ç—Ä–æ–∫–∏
        data_list = []
        
        for i in range(len(df_raw)):
            row = {}
            for col in df_raw.columns:
                val = df_raw.iloc[i][col]
                
                # –ï—Å–ª–∏ —ç—Ç–æ CLOB - —á–∏—Ç–∞–µ–º
                if hasattr(val, 'read'):
                    try:
                        content = val.read()
                        row[col] = content if content else ""
                    except:
                        row[col] = ""
                else:
                    row[col] = val if val is not None else ""
            
            data_list.append(row)
        
        # –°–æ–∑–¥–∞–µ–º –Ω–æ–≤—ã–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º
        new_df = pd.DataFrame(data_list)
        
        # –°–æ—Ö—Ä–∞–Ω—è–µ–º
        new_df.to_excel(filename, index=False, engine='openpyxl')
        
        print(f"‚úÖ –§–∞–π–ª —Å–æ—Ö—Ä–∞–Ω–µ–Ω: {filename}")
        return True
        
    except Exception as e:
        print(f"‚ùå –û—à–∏–±–∫–∞: {e}")
        return False

# –ì–õ–ê–í–ù–ê–Ø –§–£–ù–ö–¶–ò–Ø - –ò–°–ü–û–õ–¨–ó–£–ô–¢–ï –≠–¢–£
def save_with_russian_support(df_raw, filename="russian_data.xlsx"):
    """
    –ì–ª–∞–≤–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è —Å –ø–æ–¥–¥–µ—Ä–∂–∫–æ–π —Ä—É—Å—Å–∫–æ–≥–æ —è–∑—ã–∫–∞
    """
    methods = [
        ("–ü—Ä–æ—Å—Ç–æ–µ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ", simple_russian_save),
        ("–£–ª—å—Ç–∏–º–∞—Ç–∏–≤–Ω–æ–µ —Ä–µ—à–µ–Ω–∏–µ", ultimate_save_solution),
        ("–ß–µ—Ä–µ–∑ CSV", lambda df, f: save_as_csv_first(df, "temp.csv", f)),
        ("–ò—Å–ø—Ä–∞–≤–ª–µ–Ω–∏–µ –∫–æ–¥–∏—Ä–æ–≤–∫–∏", save_with_encoding_fix)
    ]
    
    for method_name, method_func in methods:
        print(f"\nüéØ –ü—Ä–æ–±—É—é: {method_name}")
        try:
            success = method_func(df_raw, filename)
            if success:
                print(f"‚úÖ {method_name} - –£–°–ü–ï–•!")
                
                # –ü—Ä–æ–≤–µ—Ä—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
                try:
                    test_df = pd.read_excel(filename, engine='openpyxl', nrows=5)
                    print("üìã –ü–µ—Ä–≤—ã–µ 5 —Å—Ç—Ä–æ–∫ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–Ω–æ–≥–æ —Ñ–∞–π–ª–∞:")
                    for col in test_df.columns:
                        sample = test_df[col].iloc[0] if len(test_df) > 0 else "N/A"
                        sample_str = str(sample)
                        # –ü—Ä–æ–≤–µ—Ä—è–µ–º –Ω–∞ –Ω–∞–ª–∏—á–∏–µ —Ä—É—Å—Å–∫–∏—Ö –±—É–∫–≤
                        russian_chars = any(char in sample_str for char in '–∞–±–≤–≥–¥–µ—ë–∂–∑–∏–π–∫–ª–º–Ω–æ–ø—Ä—Å—Ç—É—Ñ—Ö—Ü—á—à—â—ä—ã—å—ç—é—è')
                        status = "‚úÖ –ï—Å—Ç—å —Ä—É—Å—Å–∫–∏–µ" if russian_chars else "‚ö†Ô∏è –ù–µ—Ç —Ä—É—Å—Å–∫–∏—Ö"
                        print(f"   {col}: {sample_str[:30]}... {status}")
                except Exception as e:
                    print(f"‚ö†Ô∏è –ù–µ —É–¥–∞–ª–æ—Å—å –ø—Ä–æ–≤–µ—Ä–∏—Ç—å —Ñ–∞–π–ª: {e}")
                
                return True
        except Exception as e:
            print(f"‚ùå {method_name} –Ω–µ —Å—Ä–∞–±–æ—Ç–∞–ª: {e}")
            continue
    
    print("üí• –í—Å–µ –º–µ—Ç–æ–¥—ã –Ω–µ —Å—Ä–∞–±–æ—Ç–∞–ª–∏!")
    return False

# –°–ê–ú–´–ô –ü–†–û–°–¢–û–ô –í–ê–†–ò–ê–ù–¢ - –µ—Å–ª–∏ –Ω–∏—á–µ–≥–æ –Ω–µ —Ä–∞–±–æ—Ç–∞–µ—Ç
def emergency_russian_save(df_raw, filename="emergency_data.xlsx"):
    """
    –ê–≤–∞—Ä–∏–π–Ω–æ–µ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —Å –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –ø—Ä–æ—Å—Ç—ã–º –ø–æ–¥—Ö–æ–¥–æ–º
    """
    print("üö® –ê–í–ê–†–ò–ô–ù–û–ï –°–û–•–†–ê–ù–ï–ù–ò–ï")
    
    try:
        # –ü—Ä–æ—Å—Ç–æ –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –≤—Å–µ –≤ —Å—Ç—Ä–æ–∫–∏
        data = []
        for i in range(len(df_raw)):
            row = {}
            for col in df_raw.columns:
                try:
                    value = df_raw.iloc[i][col]
                    if hasattr(value, 'read'):
                        row[col] = value.read() if value else ""
                    else:
                        row[col] = str(value) if value is not None else ""
                except:
                    row[col] = ""
            data.append(row)
        
        pd.DataFrame(data).to_excel(filename, index=False, engine='openpyxl')
        print(f"‚úÖ –ê–≤–∞—Ä–∏–π–Ω–æ–µ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –∑–∞–≤–µ—Ä—à–µ–Ω–æ: {filename}")
        return True
    except Exception as e:
        print(f"‚ùå –ê–≤–∞—Ä–∏–π–Ω–æ–µ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –Ω–µ —É–¥–∞–ª–æ—Å—å: {e}")
        return False
```

```python
import pandas as pd
import numpy as np

def find_empty_columns(df):
    """
    –ù–∞—Ö–æ–¥–∏—Ç –ø–æ–ª–Ω–æ—Å—Ç—å—é –ø—É—Å—Ç—ã–µ –∫–æ–ª–æ–Ω–∫–∏ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ
    """
    # –°–ø–æ—Å–æ–± 1: –ö–æ–ª–æ–Ω–∫–∏ –≥–¥–µ –≤—Å–µ –∑–Ω–∞—á–µ–Ω–∏—è NaN/None
    completely_empty = df.columns[df.isnull().all()].tolist()
    
    # –°–ø–æ—Å–æ–± 2: –ö–æ–ª–æ–Ω–∫–∏ –≥–¥–µ –≤—Å–µ –∑–Ω–∞—á–µ–Ω–∏—è –ø—É—Å—Ç—ã–µ —Å—Ç—Ä–æ–∫–∏ (–µ—Å–ª–∏ —Ç–∏–ø object)
    empty_strings = []
    for col in df.columns:
        if df[col].dtype == 'object':
            if df[col].astype(str).str.strip().eq('').all():
                empty_strings.append(col)
    
    # –°–ø–æ—Å–æ–± 3: –ö–æ–ª–æ–Ω–∫–∏ —Å –Ω—É–ª–µ–≤—ã–º —Ä–∞–∑–Ω–æ–æ–±—Ä–∞–∑–∏–µ–º (–≤—Å–µ –∑–Ω–∞—á–µ–Ω–∏—è –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ –∏ –ø—É—Å—Ç—ã–µ)
    zero_diversity = []
    for col in df.columns:
        unique_count = df[col].nunique(dropna=False)
        if unique_count <= 1:  # 0 –∏–ª–∏ 1 —É–Ω–∏–∫–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ
            sample = df[col].iloc[0] if len(df) > 0 else None
            if pd.isna(sample) or sample == '' or sample is None:
                zero_diversity.append(col)
    
    # –û–±—ä–µ–¥–∏–Ω—è–µ–º –≤—Å–µ –ø—É—Å—Ç—ã–µ –∫–æ–ª–æ–Ω–∫–∏
    all_empty = list(set(completely_empty + empty_strings + zero_diversity))
    
    return all_empty

def analyze_column_completeness(df):
    """
    –î–µ—Ç–∞–ª—å–Ω—ã–π –∞–Ω–∞–ª–∏–∑ –∑–∞–ø–æ–ª–Ω–µ–Ω–Ω–æ—Å—Ç–∏ –∫–æ–ª–æ–Ω–æ–∫
    """
    print("üìä –ê–ù–ê–õ–ò–ó –ó–ê–ü–û–õ–ù–ï–ù–ù–û–°–¢–ò –ö–û–õ–û–ù–û–ö")
    print("=" * 50)
    
    results = []
    
    for col in df.columns:
        total_rows = len(df)
        
        # –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—É—Å—Ç—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π (NaN/None)
        null_count = df[col].isnull().sum()
        
        # –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—É—Å—Ç—ã—Ö —Å—Ç—Ä–æ–∫ (–µ—Å–ª–∏ —Ç–∏–ø object)
        empty_string_count = 0
        if df[col].dtype == 'object':
            empty_string_count = (df[col].astype(str).str.strip() == '').sum()
        
        # –û–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ "–ø—É—Å—Ç—ã—Ö" –∑–Ω–∞—á–µ–Ω–∏–π
        total_empty = null_count + empty_string_count
        
        # –ü—Ä–æ—Ü–µ–Ω—Ç –∑–∞–ø–æ–ª–Ω–µ–Ω–Ω–æ—Å—Ç–∏
        fill_percentage = ((total_rows - total_empty) / total_rows * 100) if total_rows > 0 else 0
        
        # –°—Ç–∞—Ç—É—Å –∫–æ–ª–æ–Ω–∫–∏
        if total_empty == total_rows:
            status = "üö´ –ü–û–õ–ù–û–°–¢–¨–Æ –ü–£–°–¢–ê–Ø"
        elif fill_percentage < 10:
            status = "üî¥ –û–ß–ï–ù–¨ –ú–ê–õ–û –î–ê–ù–ù–´–•"
        elif fill_percentage < 50:
            status = "üü° –ú–ê–õ–û –î–ê–ù–ù–´–•"
        else:
            status = "‚úÖ –ù–û–†–ú–ê"
        
        results.append({
            '–ö–æ–ª–æ–Ω–∫–∞': col,
            '–¢–∏–ø –¥–∞–Ω–Ω—ã—Ö': df[col].dtype,
            '–í—Å–µ–≥–æ —Å—Ç—Ä–æ–∫': total_rows,
            '–ü—É—Å—Ç—ã—Ö (NaN)': null_count,
            '–ü—É—Å—Ç—ã—Ö —Å—Ç—Ä–æ–∫': empty_string_count,
            '–í—Å–µ–≥–æ –ø—É—Å—Ç—ã—Ö': total_empty,
            '–ó–∞–ø–æ–ª–Ω–µ–Ω–æ %': round(fill_percentage, 1),
            '–°—Ç–∞—Ç—É—Å': status
        })
    
    # –°–æ–∑–¥–∞–µ–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º–∏
    analysis_df = pd.DataFrame(results)
    
    # –°–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ –ø—Ä–æ—Ü–µ–Ω—Ç—É –∑–∞–ø–æ–ª–Ω–µ–Ω–Ω–æ—Å—Ç–∏
    analysis_df = analysis_df.sort_values('–ó–∞–ø–æ–ª–Ω–µ–Ω–æ %')
    
    return analysis_df

def find_columns_with_few_values(df, threshold=1):
    """
    –ù–∞—Ö–æ–¥–∏—Ç –∫–æ–ª–æ–Ω–∫–∏ —Å –º–∞–ª—ã–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
    """
    low_diversity_cols = []
    
    for col in df.columns:
        unique_count = df[col].nunique(dropna=False)
        if unique_count <= threshold:
            # –ü—Ä–æ–≤–µ—Ä—è–µ–º, –Ω–µ —è–≤–ª—è—é—Ç—Å—è –ª–∏ –≤—Å–µ –∑–Ω–∞—á–µ–Ω–∏—è –ø—É—Å—Ç—ã–º–∏
            sample_values = df[col].dropna().unique()
            if len(sample_values) == 0 or (len(sample_values) == 1 and sample_values[0] in ['', ' ', None]):
                low_diversity_cols.append((col, unique_count))
    
    return low_diversity_cols

def quick_empty_check(df):
    """
    –ë—ã—Å—Ç—Ä–∞—è –ø—Ä–æ–≤–µ—Ä–∫–∞ –ø—É—Å—Ç—ã—Ö –∫–æ–ª–æ–Ω–æ–∫
    """
    print("üîç –ë–´–°–¢–†–ê–Ø –ü–†–û–í–ï–†–ö–ê –ü–£–°–¢–´–• –ö–û–õ–û–ù–û–ö")
    
    # –ö–æ–ª–æ–Ω–∫–∏ –ø–æ–ª–Ω–æ—Å—Ç—å—é —Å–æ—Å—Ç–æ—è—â–∏–µ –∏–∑ NaN
    nan_empty = df.columns[df.isna().all()].tolist()
    
    # –ö–æ–ª–æ–Ω–∫–∏ –ø–æ–ª–Ω–æ—Å—Ç—å—é —Å–æ—Å—Ç–æ—è—â–∏–µ –∏–∑ –ø—É—Å—Ç—ã—Ö —Å—Ç—Ä–æ–∫
    string_empty = []
    for col in df.select_dtypes(include=['object']).columns:
        if (df[col].astype(str).str.strip() == '').all():
            string_empty.append(col)
    
    # –ö–æ–ª–æ–Ω–∫–∏ —Å –æ–¥–Ω–∏–º —É–Ω–∏–∫–∞–ª—å–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º (–∫–æ—Ç–æ—Ä–æ–µ –ø—É—Å—Ç–æ–µ)
    single_value_empty = []
    for col in df.columns:
        if df[col].nunique() == 1:
            unique_val = df[col].iloc[0]
            if pd.isna(unique_val) or unique_val == '':
                single_value_empty.append(col)
    
    all_empty = list(set(nan_empty + string_empty + single_value_empty))
    
    print(f"‚úÖ –ü–æ–ª–Ω–æ—Å—Ç—å—é –ø—É—Å—Ç—ã–µ –∫–æ–ª–æ–Ω–∫–∏: {len(all_empty)}")
    if all_empty:
        print("üìã –°–ø–∏—Å–æ–∫ –ø—É—Å—Ç—ã—Ö –∫–æ–ª–æ–Ω–æ–∫:")
        for col in all_empty:
            print(f"   - {col}")
    
    return all_empty



# –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –ø–æ–ª–µ–∑–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏:

def get_column_stats(df):
    """
    –ü–æ–ª–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ø–æ –≤—Å–µ–º –∫–æ–ª–æ–Ω–∫–∞–º
    """
    stats = []
    
    for col in df.columns:
        col_stats = {
            '–ö–æ–ª–æ–Ω–∫–∞': col,
            '–¢–∏–ø': df[col].dtype,
            '–í—Å–µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏–π': len(df),
            '–ù–µ –ø—É—Å—Ç—ã—Ö': df[col].count(),
            '–ü—É—Å—Ç—ã—Ö (NaN)': df[col].isnull().sum(),
            '–£–Ω–∏–∫–∞–ª—å–Ω—ã—Ö': df[col].nunique(),
            '–ó–∞–ø–æ–ª–Ω–µ–Ω–æ %': round((df[col].count() / len(df)) * 100, 1)
        }
        
        # –ü—Ä–∏–º–µ—Ä—ã –∑–Ω–∞—á–µ–Ω–∏–π
        non_null_values = df[col].dropna()
        if len(non_null_values) > 0:
            col_stats['–ü—Ä–∏–º–µ—Ä 1'] = non_null_values.iloc[0]
            if len(non_null_values) > 1:
                col_stats['–ü—Ä–∏–º–µ—Ä 2'] = non_null_values.iloc[1]
        
        stats.append(col_stats)
    
    return pd.DataFrame(stats)

def visualize_empty_columns(df):
    """
    –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –ø—É—Å—Ç—ã—Ö –∫–æ–ª–æ–Ω–æ–∫
    """
    import matplotlib.pyplot as plt
    
    empty_percentage = (df.isnull().sum() / len(df)) * 100
    empty_percentage = empty_percentage.sort_values(ascending=False)
    
    plt.figure(figsize=(12, 6))
    empty_percentage.head(20).plot(kind='bar')  # –¢–æ–ø-20 —Å–∞–º—ã—Ö –ø—É—Å—Ç—ã—Ö
    plt.title('–ü—Ä–æ—Ü–µ–Ω—Ç –ø—É—Å—Ç—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ –∫–æ–ª–æ–Ω–∫–∞–º (–¢–æ–ø-20)')
    plt.ylabel('–ü—Ä–æ—Ü–µ–Ω—Ç –ø—É—Å—Ç—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
    
    # –ü–æ–ª–Ω–æ—Å—Ç—å—é –ø—É—Å—Ç—ã–µ –∫–æ–ª–æ–Ω–∫–∏
    completely_empty = empty_percentage[empty_percentage == 100]
    if len(completely_empty) > 0:
        print(f"üö´ –ü–æ–ª–Ω–æ—Å—Ç—å—é –ø—É—Å—Ç—ã–µ –∫–æ–ª–æ–Ω–∫–∏ ({len(completely_empty)}):")
        for col in completely_empty.index:
            print(f"   - {col}")


```

```python
# –ü—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –≤—Å–µ—Ö —Ñ—É–Ω–∫—Ü–∏–π:
if __name__ == "__main__":
    # 1. –ë—ã—Å—Ç—Ä–∞—è –ø—Ä–æ–≤–µ—Ä–∫–∞
    print("1. –ë–´–°–¢–†–ê–Ø –ü–†–û–í–ï–†–ö–ê:")
    empty_cols = quick_empty_check(df_modified)
    
    # 2. –î–µ—Ç–∞–ª—å–Ω—ã–π –∞–Ω–∞–ª–∏–∑
    print("\n2. –î–ï–¢–ê–õ–¨–ù–´–ô –ê–ù–ê–õ–ò–ó:")
    analysis_df = analyze_column_completeness(df_modified)
    print(analysis_df.to_string(index=False))
    
    # 3. –ü–æ–ª–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞
    print("\n3. –ü–û–õ–ù–ê–Ø –°–¢–ê–¢–ò–°–¢–ò–ö–ê:")
    stats_df = get_column_stats(df_modified)
    print(stats_df.to_string(index=False))
    
    # 4. –ö–æ–ª–æ–Ω–∫–∏ –¥–ª—è —É–¥–∞–ª–µ–Ω–∏—è (–ø–æ–ª–Ω–æ—Å—Ç—å—é –ø—É—Å—Ç—ã–µ)
    if empty_cols:
        print(f"\nüóëÔ∏è  –†–µ–∫–æ–º–µ–Ω–¥—É–µ—Ç—Å—è —É–¥–∞–ª–∏—Ç—å {len(empty_cols)} –ø—É—Å—Ç—ã—Ö –∫–æ–ª–æ–Ω–æ–∫:")
        for col in empty_cols:
            print(f"   - {col}")
        
        # –£–¥–∞–ª–µ–Ω–∏–µ –ø—É—Å—Ç—ã—Ö –∫–æ–ª–æ–Ω–æ–∫ (–æ–ø—Ü–∏–æ–Ω–∞–ª—å–Ω–æ)
        # df_clean = df_raw.drop(columns=empty_cols)
        # print(f"‚úÖ –£–¥–∞–ª–µ–Ω–æ {len(empty_cols)} –ø—É—Å—Ç—ã—Ö –∫–æ–ª–æ–Ω–æ–∫")
```
